```python
import pandas as pd

# Cargar los datos de cada tienda
tienda_1 = pd.read_csv('tienda_1.csv')
tienda_2 = pd.read_csv('tienda_2.csv')
tienda_3 = pd.read_csv('tienda_3.csv')
tienda_4 = pd.read_csv('tienda_4.csv')

# Combinar los DataFrames en uno solo
df = pd.concat([tienda_1, tienda_2, tienda_3, tienda_4], ignore_index=True)

# Verifica que los datos se hayan cargado correctamente
print(df.head())
```

```python
# Verificar los nombres de las columnas en el DataFrame
print(df.columns)
```

```python
import pandas as pd
import matplotlib.pyplot as plt  # Asegúrate de importar matplotlib

# Crear una nueva columna de ventas
df['Ventas'] = df['Precio'] * df['Cantidad de cuotas']

# Agrupar por producto y sumar las ventas
ventas_por_producto = df.groupby('Producto')['Ventas'].sum().reset_index()

# Ordenar de mayor a menor
ventas_por_producto = ventas_por_producto.sort_values(by='Ventas', ascending=False)

# Seleccionar los 10 productos más vendidos
top_10_productos = ventas_por_producto.head(10)

# Crear el gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(top_10_productos['Producto'], top_10_productos['Ventas'], color='blue')
plt.title('Productos Más Vendidos')
plt.xlabel('Productos')
plt.ylabel('Ventas')
plt.xticks(rotation=45)
plt.show()
```

```python
# Convertir la columna 'Fecha de Compra' a tipo datetime
df['Fecha de Compra'] = pd.to_datetime(df['Fecha de Compra'])
```

```python
# Crear una nueva columna de ventas si no lo has hecho
df['Ventas'] = df['Precio'] * df['Cantidad de cuotas']

# Agrupar por mes y sumar las ventas
ventas_por_mes = df.resample('M', on='Fecha de Compra')['Ventas'].sum().reset_index()
```

```python
# Crear el gráfico de líneas
plt.figure(figsize=(10, 6))
plt.plot(ventas_por_mes['Fecha de Compra'], ventas_por_mes['Ventas'], marker='o', color='orange')
plt.title('Ventas Totales por Mes')
plt.xlabel('Mes')
plt.ylabel('Ventas')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
# Crear el gráfico de dispersión
plt.figure(figsize=(10, 6))
plt.scatter(df['Precio'], df['Cantidad de cuotas'], alpha=0.5, color='green')
plt.title('Relación entre Precio y Cantidad Vendida')
plt.xlabel('Precio')
plt.ylabel('Cantidad Vendida')
plt.grid()
plt.show()
```

2. Correlación entre Ventas y Calificaciones
Al correlacionar las ventas con las calificaciones promedio de los clientes, se observa que las tiendas con mayores ventas, como Bogotá (calificación de 4.11) y Cali (calificación de 4.05), también tienen calificaciones relativamente altas. Esto indica que la satisfacción del cliente puede estar relacionada con el rendimiento de ventas, sugiriendo que las tiendas que ofrecen una mejor experiencia al cliente tienden a tener un mejor desempeño en ventas.

3. Factores Externos
Varios factores externos pueden influir en las ventas observadas:

Densidad de Población: Las áreas con mayor densidad de población, como Bogotá y Medellín, tienden a tener más clientes potenciales, lo que podría explicar sus altos ingresos.
Competencia: En áreas con alta competencia, como Barranquilla (4,373,335,000), las ventas pueden verse afectadas negativamente. Es importante investigar cómo se posicionan las tiendas en relación con sus competidores.
Accesibilidad: La facilidad de acceso a las tiendas también puede influir en las ventas. Las tiendas ubicadas en áreas de fácil acceso pueden atraer a más clientes.
4. Recomendaciones
Basado en el análisis, se recomienda que el Sr. Juan considere expandir sus operaciones en áreas con alta densidad poblacional y buenas calificaciones, como Bogotá y Cali.
