```python
import pandas as pd

# Carga de datos
url = 'https://raw.githubusercontent.com/alura-cursos/challenge2-data-science-LATAM/main/TelecomX_Data.json'
df = pd.read_json(url)
# Visualizar las primeras filas
df.head()
```

```python
# Ver información general de las columnas
df.info()
```

```python
# Desglosar columnas anidadas
df_normalizado = pd.json_normalize(df.to_dict(orient='records'), sep='_')

# Ver las nuevas columnas creadas
df_normalizado.head()
```

```python
# Verificar valores en la columna Churn
df_normalizado['Churn'].value_counts(dropna=False)
```

```python
# Eliminar filas con Churn vacío
df_limpio = df_normalizado[df_normalizado['Churn'] != ""]

# Verificar que ya no haya vacíos
df_limpio['Churn'].value_counts()
```

```python
# Crear copia y convertir la columna correcta
df_final = df_limpio.copy()

# Conversión a número
df_final['account_Charges_Total'] = pd.to_numeric(df_final['account_Charges_Total'], errors='coerce')

# Verificar nulos
print(f"Nulos en Total: {df_final['account_Charges_Total'].isnull().sum()}")
```

```python
# Eliminar las 11 filas con nulos
df_final = df_final.dropna(subset=['account_Charges_Total'])

# Confirmar limpieza total (debe dar 0)
print(f"Nulos restantes: {df_final['account_Charges_Total'].isnull().sum()}")
```
