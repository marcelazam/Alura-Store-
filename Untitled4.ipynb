```python
import pandas as pd

# Carga de datos
url = 'https://raw.githubusercontent.com/alura-cursos/challenge2-data-science-LATAM/main/TelecomX_Data.json'
df = pd.read_json(url)
# Visualizar las primeras filas
df.head()
```

```python
customer_df = pd.json_normalize(df['customer'])
customer_df.head()
```

```python
phone_df = pd.json_normalize(df['phone'])
phone_df.head()
```

```python
internet_df = pd.json_normalize(df['internet'])
internet_df.head()
```

```python
account_df = pd.json_normalize(df['account'])
account_df.head()
```

```python
df_final = pd.concat(
    [
        df[['customerID', 'Churn']],
        customer_df,
        phone_df,
        internet_df,
        account_df
    ],
    axis=1
)

df_final.head()
```

```python
df_final.info()
```

```python
df_final['Charges.Total'] = pd.to_numeric(df_final['Charges.Total'], errors='coerce')

# Revisar que se hizo correctamente
df_final['Charges.Total'].dtype, df_final['Charges.Total'].isna().sum()
```

```python
# Estadsticas descriptivas para columnas numricas
desc = df_final.describe().T
desc
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Contar clientes por Churn
churn_counts = df_final['Churn'].value_counts()

# Mostrar proporciones
print(churn_counts)

# Grfico de barras
plt.figure(figsize=(6,4))
sns.barplot(x=churn_counts.index, y=churn_counts.values, palette='Set2')
plt.title('Distribucin de Churn (evasin)')
plt.ylabel('Nmero de clientes')
plt.xlabel('Churn')
plt.show()
```

### Distribucin de evasin (Churn)

- La mayora de los clientes **permanecieron**: 5174 (~73%)  
- Los clientes que **se dieron de baja** fueron 1869 (~27%)  
- Esto muestra que la evasin es significativa pero no mayoritaria, lo que permite enfocarnos en estrategias para reducirla.

El grfico de barras a continuacin visualiza claramente estas proporciones.


```python
# Lista de variables categricas a analizar
categorical_vars = ['gender', 'Partner', 'Dependents', 'PhoneService',
                    'MultipleLines', 'InternetService', 'OnlineSecurity',
                    'OnlineBackup', 'DeviceProtection', 'TechSupport',
                    'StreamingTV', 'StreamingMovies', 'Contract',
                    'PaperlessBilling', 'PaymentMethod']

# Recuento de Churn por cada variable categrica
for var in categorical_vars:
    print(f"=== Churn por {var} ===")
    print(pd.crosstab(df_final[var], df_final['Churn']))
    print("\n")
```

#### Conteo de evasin por variables numricas

Se analizaron las variables numricas `tenure`, `Charges.Monthly` y `Charges.Total` para identificar patrones de evasin (`Churn`):

- **Tenure (meses de permanencia)**:
  - Los clientes que abandonan (`Yes`) tienden a tener **menos meses de permanencia**, mientras que los que permanecen (`No`) tienen mayor tenure.
  - Esto indica que los clientes nuevos o con menor tiempo en la empresa tienen mayor riesgo de churn.

- **Charges.Monthly (cobro mensual)**:
  - Los clientes que abandonan tienden a pagar **ms o menos que el promedio**, mostrando cierta dispersin mayor.
  - Esto sugiere que cargos altos o inestables pueden influir en la decisin de cancelar.

- **Charges.Total (cobro total acumulado)**:
  - Los clientes que permanecen tienen **mayor total gastado**, mientras que los que abandonan suelen tener menor gasto total.
  - Esto refleja que la retencin est asociada con el tiempo de permanencia y gasto acumulado.

**Interpretacin general:**
- Variables numricas como **tenure y total gastado** son buenos indicadores de riesgo de churn.
- Esta informacin es til para disear estrategias de fidelizacin, enfocndose en clientes con bajo tenure o bajo gasto acumulado.
