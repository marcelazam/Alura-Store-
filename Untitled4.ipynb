```python
import pandas as pd

# Carga de datos
url = 'https://raw.githubusercontent.com/alura-cursos/challenge2-data-science-LATAM/main/TelecomX_Data.json'
df = pd.read_json(url)
# Visualizar las primeras filas
df.head()
```

```python
customer_df = pd.json_normalize(df['customer'])
customer_df.head()
```

```python
phone_df = pd.json_normalize(df['phone'])
phone_df.head()
```

```python
internet_df = pd.json_normalize(df['internet'])
internet_df.head()
```

```python
account_df = pd.json_normalize(df['account'])
account_df.head()
```

```python
df_final = pd.concat(
    [
        df[['customerID', 'Churn']],
        customer_df,
        phone_df,
        internet_df,
        account_df
    ],
    axis=1
)

df_final.head()
```

```python
df_final.info()
```

```python
df_final['Charges.Total'] = pd.to_numeric(df_final['Charges.Total'], errors='coerce')

# Revisar que se hizo correctamente
df_final['Charges.Total'].dtype, df_final['Charges.Total'].isna().sum()
```
